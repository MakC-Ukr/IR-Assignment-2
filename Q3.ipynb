```python
path = "/content/drive/MyDrive/IR/IR-assignment-2-data (2).txt"

File_data = np.loadtxt(path, dtype='str')

arr = File_data
newarr = []
for i in range(len(arr)):
  if(arr[i][1]=="qid:4"):
    newarr.append(arr[i])

relevencescore = []
dict = {}
for j in range(len(newarr)):
  for k in range(2,len(newarr[0])):
    x=newarr[j][k].split(":")
    if j not in dict:
      dict[j] = {}
    if j in dict:
      if int(x[0]) not in dict[j].keys():
        dict[j][int(x[0])] = float(x[1])

for i in range(len(dict)):
  relevencescore.append([])
  for j in range(1,len(dict[i])+1):
    relevencescore[i].append(dict[i][j])


print(dict)
print(relevencescore)

```

```python
def dcg(relevance_scores):
    discount = lambda i: 1 / np.log2(i + 2)
    cum_gain = sum([score * discount(i) for i, score in enumerate(relevance_scores)])
    return cum_gain

def max_dcg(relevance_scores):
    sorted_scores = sorted(relevance_scores, reverse=True)
    max_dcg = dcg(sorted_scores)
    return max_dcg
```

```python
dcg_score = []
max_dcg_score = []
for i in range(len(relevencescore)):
  dcg_score.append(dcg(relevencescore[i]))
  max_dcg_score.append(max_dcg(relevencescore[i]))

print(dcg_score)
print(max_dcg_score)
```

```python
def ndcg(relevance_scores, k=None):
    if k is None:
        k = len(relevance_scores)
    max_dcg = dcg(sorted(relevance_scores, reverse=True)[:k])
    if max_dcg == 0:
        return 0
    actual_dcg = dcg(relevance_scores[:k])
    ndcg = actual_dcg / max_dcg
    return ndcg
```

```python
ndcg_score = []
for i in range(len(relevencescore)):
  ndcg_score.append(ndcg(relevencescore[i],50))

print(ndcg_score)
```
